## Cleaning Process

```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer

nltk.download('punkt')
nltk.download('stopwords')

def clean_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    return text.strip()

def clean_paragraph(paragraph):
    # Tokenize into sentences
    sentences = sent_tokenize(paragraph)

    # Clean each sentence
    cleaned_sentences = [clean_text(sentence) for sentence in sentences]

    # Join the sentences back into a paragraph
    cleaned_paragraph = ' '.join(cleaned_sentences)

    return cleaned_paragraph
```

### remove title from pargrapgh text

```python
parDf["pargragh"]=[par.replace(title,"") for par,title in zip(parDf["pargragh"].to_list(),parDf["title"].to_list())]
```

### apply cleaning of both

```python
parDf["pargragh"]=parDf["pargragh"].apply(lambda x:clean_paragraph(x))
parDf["title"]=parDf["title"].apply(lambda x:clean_paragraph(x))
```

```python
train,test = train_test_split(parDf,test_size=0.1)
validation,test = train_test_split(test,test_size=0.05)

dstrain=Dataset.from_pandas(train)
dsval=Dataset.from_pandas(validation)
dstest=Dataset.from_pandas(test)
ds = DatasetDict()
ds['train'] = dstrain
ds['validation'] = dsval
ds["test"]=dstest
```

```python
tokenizer.pad_token=tokenizer.eos_token
```

```python
def tokenize(element):
    outputs = tokenizer(
        element['pargragh']
        ,padding=True,return_attention_mask=True
    )
    outputs2 = tokenizer(
        element['title']
        ,padding=True,return_attention_mask=True
    )
    return {"input_ids": outputs["input_ids"],"labels":outputs2["input_ids"]}
```

```python
tokenized_datasets = ds.map(
    tokenize, batched=True, remove_columns=ds["train"].column_names
)
tokenized_datasets

```

```python

# from transformers import DataCollatorForLanguageModeling
data_collator = DataCollatorForSeq2Seq(tokenizer)
```

```python
from datasets import load_metric
metric = load_metric("rouge")
```

```python
metric
```

```python
import numpy as np
import nltk
import torch, gc
nltk.download('punkt')


def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    decoded_preds=[tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens([x.argmax() for x in elem], skip_special_tokens=True)) for elem in predictions[0] ]
    decoded_labels=[tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(elem, skip_special_tokens=True)) for elem in labels]
    # Compute ROUGE scores
    result = metric.compute(predictions=decoded_preds, references=decoded_labels)
    # Extract ROUGE f1 scores
    result = {key: value.mid.fmeasure * 100 for key, value in result.items()}

    # Add mean generated length to metrics
    prediction_lens = [np.count_nonzero(pred != tokenizer.pad_token_id)
                      for pred in predictions]
    result["gen_len"] = np.mean(prediction_lens)

    return {k: round(v, 4) for k, v in result.items()}
```

```python
trainer = Trainer(
    model=model,
    args=args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    data_collator=data_collator
    ,compute_metrics=compute_metrics
)
```

```python
parDf["tokensTitleLen"]=parDf["title"].apply(lambda x:len(tokenizer(x)["input_ids"]))
```

```python
def prediction(text):
    preds=tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(model.generate(inputs=tokenizer(text,return_tensors="pt")["input_ids"],max_new_tokens=parDf["tokensTitleLen"].max())[0].tolist(),skip_special_tokens=True))
    return preds
```

```python
test["preds"]=test["pargragh"].apply(lambda x:prediction(x))
```

```python
rouge1score=[]
for pred,real in zip(test["preds"],test["title"]):
    result=metric.compute(predictions=[pred], references=[real])
    result = {key: value.mid.fmeasure * 100 for key, value in result.items()}["rouge1"]
    rouge1score.append(result)
```
